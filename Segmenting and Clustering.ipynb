```python
df_pandas = pd.read_html(html_tab) #converting to pandas to arrange.
df=df_pandas[0]
```

```python
#Importing geo positions from the list provided in the assignment
df_geo = pd.read_csv('https://cocl.us/Geospatial_data')
df_geo
```

I have the table but what I see is the table column title has the different name for post code. So I will arrange the df_geo first.

```python
df_geo.rename(columns={'Postal Code':'Post Code'},inplace=True)
df_geo.head()
```

Merging two tables.

```python
df_merged = pd.merge(df_dropped,df_geo,on='Post Code')
df_merged
```

The analysis is for Toronto then I should take only the cells includes Toronto. 

```python
df_toronto = df_merged[df_merged['Borough'].str.contains('Toronto',regex=False)]
df_toronto.shape
```
